```python
data["time"] = pd.to_datetime(data["time"])
data.rename(columns={
    "time": "Date",
    "relativehumidity_2m": "Humidity Inst. (%)",
    "apparent_temperature": "Air Temp. Inst. (Â°C)",
    "windspeed_10m": "Wind Speed 10 m Syno. (km/h)",
    "winddirection_10m": "Wind Dir. 10 m Syno. (Â°)",
    "precipitation": "Precip. (mm)",
    "shortwave_radiation": "Incoming Solar Rad. (W/m2)",
    }, inplace=True)

columns = [
    "Date",
    "Air Temp. Inst. (Â°C)",
    "Humidity Inst. (%)",
    "Wind Speed 10 m Syno. (km/h)",
    "Wind Dir. 10 m Syno. (Â°)",
    "Precip. (mm)",
    "Incoming Solar Rad. (W/m2)"
]

columns_to_drop = [column for column in data.columns if column not in columns]
data.drop(columns=columns_to_drop, axis=1, inplace=True)

data["Wind Speed 10 m Avg. (km/h)"] = data["Wind Speed 10 m Syno. (km/h)"].rolling(window=24).mean()
data["Wind Speed 10 m Avg. (km/h)"].fillna(0, inplace=True)

data["Wind Dir. 10 m Avg. (Â°)"] = data["Wind Dir. 10 m Syno. (°)"].rolling(window=24).mean()
data["Wind Dir. 10 m Avg. (Â°)"].fillna(0, inplace=True)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"
data["Year"]  = data["Date"].dt.year
data["Month"]  = data["Date"].dt.month
data["Day"]  = data["Date"].dt.day
data["Hour"]  = data["Date"].dt.hour
data.drop("Date", axis=1, inplace=True)  # Drop the original date column
data.sort_index(inplace=True)  # Ensure the data is sorted by time
data["Season"] = data["Month"].apply(get_season)
```
